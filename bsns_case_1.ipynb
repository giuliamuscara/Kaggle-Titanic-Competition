```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
campaign_tasks_df = pd.read_csv("/content/campaign_tasks.csv")
customers_df = pd.read_csv("/content/customers.csv")
transactions_df = pd.read_csv("/content/transactions.csv")
```

```
customers_df.info()
customers_df.describe()
customers_df.head(6)
```

```
# Analyze the customer dataframe
print("Number of unique customers:", customers_df['customer_id'].nunique())
print("\nDistribution of genders:")
display(customers_df['gender'].value_counts())
print("\nDistribution of countries:")
display(customers_df['country'].value_counts())
print("\nAge distribution:")
# Assuming 'birth_year' is available in customers_df
current_year = pd.to_datetime('now').year
customers_df['age'] = current_year - customers_df['birth_year']
print("Mean age:", customers_df['age'].mean())
print("Median age:", customers_df['age'].median())
print("Age range:", customers_df['age'].min(), "-", customers_df['age'].max())
```

```
transactions_df.info()
transactions_df.describe()
```

```
# Analyze the transactions dataframe
print("Number of unique transactions:", transactions_df.shape[0])
print("\nDistribution of event types:")
display(transactions_df['event_type'].value_counts())
print("\nDistribution of product categories:")
display(transactions_df['product_category'].value_counts())
print("\nQuantity distribution:")
print("Mean quantity:", transactions_df['quantity'].mean())
print("Median quantity:", transactions_df['quantity'].median())
print("Quantity range:", transactions_df['quantity'].min(), "-", transactions_df['quantity'].max())
print("\nNet sale euro distribution:")
print("Mean net sale euro:", transactions_df['net_sale_euro'].mean())
print("Median net sale euro:", transactions_df['net_sale_euro'].median())
print("Net sale euro range:", transactions_df['net_sale_euro'].min(), "-", transactions_df['net_sale_euro'].max())
```

```
# Analyze the campaign_tasks dataframe
print("Number of unique campaign tasks:", campaign_tasks_df.shape[0])
print("\nNumber of unique campaigns:", campaign_tasks_df['id_campaign'].nunique())
print("\nNumber of unique customers involved in campaigns:", campaign_tasks_df['customer_id'].nunique())
print("\nDistribution of campaign statuses:")
display(campaign_tasks_df['campaign_status'].value_counts())
print("\nDistribution of campaign channels:")
display(campaign_tasks_df['campaign_channel'].value_counts())
print("\nDistribution of activity subjects:")
display(campaign_tasks_df['activity_subject'].value_counts())
# Convert date columns to datetime objects
campaign_tasks_df['start_date'] = pd.to_datetime(campaign_tasks_df['start_date'])
campaign_tasks_df['end_date'] = pd.to_datetime(campaign_tasks_df['end_date'])
campaign_tasks_df['last_modified_date'] = pd.to_datetime(campaign_tasks_df['last_modified_date'])
print("\nCampaign duration distribution (in days):")
campaign_tasks_df['campaign_duration'] = (campaign_tasks_df['end_date'] - campaign_tasks_df['start_date']).dt.days
print("Mean duration:", campaign_tasks_df['campaign_duration'].mean())
print("Median duration:", campaign_tasks_df['campaign_duration'].median())
print("Duration range:", campaign_tasks_df['campaign_duration'].min(), "-", campaign_tasks_df['campaign_duration'].max())
```

```
# Get the list of customer_ids where is_active is 0
inactive_customer_ids = customers_df[customers_df['is_active'] == 0]['customer_id']
print("Number of inactive customers: ", len(inactive_customer_ids))

# Filter transactions_df
inactive_transactions_df = transactions_df[transactions_df['customer_id'].isin(inactive_customer_ids)]

# Filter campaign_tasks_df
inactive_campaign_tasks_df = campaign_tasks_df[campaign_tasks_df['customer_id'].isin(inactive_customer_ids)]

print("Transactions for inactive customers:")
display(inactive_transactions_df.shape)

print("\nCampaign tasks for inactive customers:")
display(inactive_campaign_tasks_df.shape)
```

```
# Convert event_date to datetime objects
transactions_df['event_date'] = pd.to_datetime(transactions_df['event_date'])

# Create a dictionary of customer IDs and their transaction dates
customer_dates = transactions_df.groupby('customer_id')['event_date'].apply(list).to_dict()

# Display the first few entries of the dictionary to verify
for customer_id, dates in list(customer_dates.items())[:5]:
    print(f"Customer ID: {customer_id}")
    print(f"Transaction Dates: {dates}")
```

```
from datetime import timedelta

count_customers_with_long_interval = 0

for customer_id, dates in customer_dates.items():
    if len(dates) > 1:
        #order dates from most recent to oldest timestamp
        dates.sort(reverse=True)
        prev_ts = dates[0]
        for ts in dates:
          if (prev_ts - ts) >= timedelta(days=120):
              count_customers_with_long_interval += 1
              break


print(f"Number of customers with transaction dates at least 120 days apart: {count_customers_with_long_interval}")
print("Probability that a customer will buy again after being inactive: ", count_customers_with_long_interval / len(customers_df['customer_id']))
```
